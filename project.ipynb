```python
import matplotlib.pyplot as plt

import pandas as pd

import seaborn as sns

data = pd.read_csv('environmental_data.csv')
df = pd.DataFrame(data)

# Plotting pairwise relationships using seaborn
sns.pairplot(df[["Temperature", "Humidity", "Pressure"]],diag_kind='kde')

# Adding a title to the plot
plt.suptitle("Pairwise Relationships Among Temperature, Humidity, and Pressure", y=1.123)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv("environmental_data.csv")

# Display the first few rows of the data
print(df.head())

# Plot 1: Line Graph for Temperature, Humidity, and Pressure
plt.figure(figsize=(10, 6))

plt.plot(df.index, df["Temperature"], label="Temperature", marker='o', color="blue")

plt.plot(df.index, df["Humidity"], label="Humidity", marker='s', color="green")

plt.plot(df.index, df["Pressure"], label="Pressure", marker='^', color="red")

plt.title("Line Graph: Temperature, Humidity, and Pressure Over Time")

plt.xlabel("Index")

plt.ylabel("Values")

plt.legend()

plt.grid(alpha=0.5)

plt.show()

#  Bar Graph for Temperature, Humidity, and Pressure
plt.figure(figsize=(10, 6))
bar_width = 0.25
x = df.index

# Create side-by-side bars
plt.bar(x - bar_width, df["Temperature"], width=bar_width, label="Temperature", color="blue")

plt.bar(x, df["Humidity"], width=bar_width, label="Humidity", color="green")

plt.bar(x + bar_width, df["Pressure"], width=bar_width, label="Pressure", color="red")

plt.title("Bar Graph: Temperature, Humidity, and Pressure")

plt.xlabel("Index")

plt.ylabel("Values")

plt.legend()

plt.grid(alpha=0.3)

plt.show()
```
